```python
import matplotlib.pyplot as plt
import numpy as np
a = [0.1, 0.24, 0.38, 0.52, 0.66, 0.80, 0.94, 1.08, 1.22, 1.36, 1.50, 1.65, 1.79, 1.93, 3.26,3.53, 3.80, 4.07, 4.34, 4.61, 15.0, 25.0, 34.0, 53.0, 62.0]
newList = [x / 1000 for x in a]
b = [11.1, 19.9, 25.4, 28.8, 31.9, 34.2, 36.3, 38.9, 41.0, 42.8, 44.4, 46.0, 46.9, 48.7, 59.0, 61.1, 62.9, 64.3, 65.6, 67.3, 106.5, 130.0, 145.0, 175.0, 185.0]
x = np.array(newList)
y= np.array(b)
x = np.log(x)
y = np.log(y)
plt.plot(x,y)
```

```python
from scipy import optimize
def func(x, a, b):
    y = a*x + b
    return y

alpha = optimize.curve_fit(func, xdata = x, ydata = y)[0]
print(alpha)
```

```python
plt.figure(figsize = (10,8))
plt.plot(x, y, 'b.',label = 'Given Data')
plt.plot(x, alpha[0]*x + alpha[1], 'r', label='Best fit model')
plt.xlabel('t(s)')
plt.ylabel('R(m)')
plt.legend()
#plt.show()
plt.savefig('linearregression.png', dpi =300)
```
